```python
import numpy as np
import pandas as pd
from qmeans.qmeans import QuantumKMeans
from qiskit import Aer
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
backend = Aer.get_backend("aer_simulator_statevector")
X,y = load_iris(return_X_y=True)
print('Data is: ',X)
print('Labels are: ',y)
```

```python
qk_means = QuantumKMeans(backend, n_clusters=3, verbose=True, map_type='probability', init='random', shots=10000)
qk_means.fit(X)
print(qk_means.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=qk_means.labels_)
```

```python
qk_means.get_params()
```

# Code version (qiskit >= 1.0.0)

```python
import numpy as np
import pandas as pd
from qmeans.qmeans import QuantumKMeans, AerSimulator
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
backend = AerSimulator()
X,y = load_iris(return_X_y=True)
print('Data is: ',X)
print('Labels are: ',y)
```

```python
qk_means = QuantumKMeans(backend, n_clusters=3, verbose=True, map_type='probability', init='random', shots=10000)
qk_means.fit(X)
print(qk_means.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=qk_means.labels_)
```

```python
qk_means.get_params()
```
